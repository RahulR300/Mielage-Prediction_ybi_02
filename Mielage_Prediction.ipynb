## **Mielage predection - Regression Analysis**


### **Title of project:** Mielage predection-regression analysis


### **Source:**
This dataset taken from StatLib Library which is maintained at Carnegie Mellon University.The dataset was used in the 1983 America statistical accosiation exposition

### **Data Set Information:**

This dataset is a slightly modified version of the dataset provided in the StatLib library in line with the use by Ross Quinlan (1993) in predicting the attribute "mpg", 8 of the original instances were removed because they had unknown values for the 'mpg' attribute. The original dataset is available in the file "auto-mpg.data-original".

"The data concerns city-cycle fuel consumption in miles per gallon, to be predicted in terms of 3 multivalued discrete and 5 continuous attributes" (Quinlan, 1993)*italicized text*

### **Attribute Information:**

1. mpg.continuous
2. cylinders: multi-valued discrete
3. displacement: continuous
4. horsepower: continuous
5. weight: continuous
6. acceleration: continuous
7. model year: multi-valued discrete
8. origin multi-valued discrete
9. car name: string (unique for each instance)


# **Import Libraries**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

# **Import Data**

```python
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

```python
df.head()
```

```python
df.nunique()
```

# **Data Preprocessing**

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

# **Removing Missing Values**

```python
df = df.dropna()
```

```python
df.info()
```

# **Data Visualization**

```python
sns.pairplot(df, x_vars=['displacement','horsepower','weight','acceleration','mpg'],y_vars=['mpg']);
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = df);
```

Define Target Variable y & feature X

```python
df.columns
```

```python
y = df['mpg']
```

```python
y.shape
```

```python
X = df[['displacement', 'horsepower', 'weight','acceleration']]
```

```python
X.shape
```

```python
X
```

# **Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

### **After Standardization Mean is Zero and Standard Deviation is One**

# **Train-Test Split Data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

Mileage = 23.4 -1.05 Displacement -1.68 Horsepower -4.10 Weight -0.115 Acceleration + error

# **Predict Test Data**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

# **Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

# **Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True,include_bias=False)
```

```python
X_train2 = poly.fit_transform(X_train)
```

```python
X_test2=poly.fit_transform(X_test)
```

```python
lr.fit(X_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly=lr.predict(X_test2)
```

# **Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test, y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test, y_pred_poly)
```

```python
r2_score(y_test, y_pred_poly)
```
